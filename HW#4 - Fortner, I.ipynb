```python
x = np.array([0, 5, 10, 20])
y = np.array([2, 4, 6, 8])

plt.subplot(2, 2, 1)
plt.plot(x,y, color = "green")
plt.title("Quiz Scores")
plt.grid(color = 'red', linestyle = '--', linewidth = 0.2)


x = np.array([0, 5, 10, 20])
y = np.array([3, 9, 12, 15])

plt.subplot(2, 2, 2)
plt.plot(x,y, color = "green")
plt.title("Test Scores")
plt.grid(color = 'red', linestyle = '--', linewidth = 0.2)



x = np.array([0, 2, 4, 6])
y = np.array([1, 3, 5, 7])

plt.subplot(2, 2, 3)
plt.plot(x,y, color = "green")
plt.title("Homework Scores")
plt.grid(color = 'red', linestyle = '--', linewidth = 0.2)


x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([15, 13, 10, 6, 2, 2])

plt.subplot(2, 2, 4)
plt.bar(x,y, width = .5, color = "darkgreen")
plt.title("Attendance")
plt.grid(color = 'red', linestyle = '--', linewidth = 0.2)
plt.xlabel("Days Missed")
plt.ylabel("Number of Students")

plt.tight_layout()
plt.suptitle("STAT 1129")
plt.show()
```
